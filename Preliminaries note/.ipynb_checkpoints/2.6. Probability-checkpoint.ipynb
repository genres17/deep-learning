```python
import torch
from torch.distributions.multinomial import Multinomial

from IPython import display

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
probabilities = torch.ones([6], dtype = torch.float64) / 6
distribution = Multinomial(total_count = 1, probs = probabilities)
distribution.sample()
```

以上是对一个所有元素取值概率都为1/6，然后对其进行一次取样，返回的是取样结果次数

```python
print((Multinomial(total_count = 10, probs = probabilities)).sample())
print((Multinomial(total_count = 100, probs = probabilities)).sample())
```

可以通过修改total_count参数来修改总取样的值

```python
counts = torch.zeros(6, 1000, dtype = torch.double)
counts
```

```python
num_rolls = 1000
rolls = Multinomial(total_count = 1, probs = probabilities)
counts = torch.zeros(6, 1000, dtype = torch.double)
totals = torch.zeros(6, dtype = torch.double)
for i in range(num_rolls):
    totals = totals + rolls.sample()
    counts[:, i] = totals
    if i<20:
        print(totals)
        print(counts)
        print('===')
```

totals所求的是模拟1000次抽取后每个样例抽到的次数向量
counts则是把每次求得的totals向量转置后记录，一共记录了1000列

```python
x = torch.arange(1000, dtype = torch.double).reshape((1,1000)) + 1
estimates = counts / x
estimates # 每列都是一组组合概率
```

```python
display.set_matplotlib_formats('svg')

plt.figure(figsize=(8, 6))
for i in range(6):
    plt.plot(estimates[i, :].numpy(), label=("P(die=" + str(i) +")"))

plt.axhline(y=0.16666, color='black', linestyle='dashed')
plt.legend()
plt.show()

```

1. 对于任意两个随机事件的并集的概率，一定小于等于其两个独立概率之和
2. 两个互斥事件的并集的概率，等于两个独立概率之和
3. P(A,B)=P(B∣A)P(A) P(A,B)=P(A∣B)P(B)
4. 另一个需要检查的有用属性是依赖性与独立性。两个随机变量A和B是独立的，意味着A的一个事件的发生不会揭示B事件发生的任何信息。在这种情况下，P（B∣A）=P（B）。统计学家通常将其表示为A⊥B。根据Bayes定理，它立即得出P（A∣B）=P（A）。在所有其他的情况下我们称之为A和B是独立事件。因为P(A∣B)=P(A,B)P(B)=P(A)可以推导出P(A,B)=P(A)P(B)，两个随机变量是独立的当且仅当它们的联合分布是各自分布的乘积。同样，两个随机变量A和B是条件独立的，当且仅当P（A，B∣C）=P（A∣C）P（B∣C）
