## 线性回归神经元操作实现

### 导包

```python
from IPython import display
from matplotlib import pyplot as plt
import torch
import random
```

### 生成数据集

我们构造一个简单的人工训练数据集，它可以使我们能够直观比较学到的参数和真实的模型参数的区别。设训练数据集样本数为1000，输入个数（特征数）为2。给定随机生成的批量样本特征X∈R1000×2，我们使用线性回归模型真实权重w=[2,−3.4]⊤和偏差b=4.2，以及一个随机噪声项ϵ

来生成标签
y=Xw+b+ϵ,

其中噪声项ϵ
服从均值为0、标准差为0.01的正态分布。噪声代表了数据集中无意义的干扰。

```python
num_inputs = 2
num_examples = 1000
true_w = torch.tensor([2, -3.4])
true_b = 4.2
features = torch.zeros(size=(num_examples, num_inputs)).normal_() #返回一个张量，包含从给定参数means,std的离散正态分布中抽取随机数
labels = torch.matmul(features, true_w) + true_b
labels += torch.zeros(size=labels.shape).normal_(std=0.01)
labels
```

features的每一行是一个长度为2的向量，而labels的每一行是一个长度为1的向量（标量）
这里补充一下matmul的乘法机制：
matmul(input,other)
1. 若两个都是1D（向量）的，则返回两个向量的点积

```python
x = torch.rand(2)
y = torch.rand(2)
print(torch.matmul(x,y),torch.matmul(x,y).size())
```

2. 若两个都是2D（矩阵）的，则按照（矩阵相乘）规则返回2D

```python
x = torch.rand(2,4)
y = torch.rand(4,3) ###维度也要对应才可以乘
print(torch.matmul(x,y),'\n',torch.matmul(x,y).size())
```

3. 若input维度1D，other维度2D，则先将1D的维度扩充到2D（1D的维数前面+1），然后得到结果后再将此维度去掉，得到的与input的维度相同。即使作扩充（广播）处理，input的维度也要和other维度做对应关系。

```python
x = torch.rand(4) #1D
y = torch.rand(4,3) #2D
print(x.size())
print(y.size())
print(torch.matmul(x,y),'\n',torch.matmul(x,y).size())
```

4. 若input是2D，other是1D，则返回两者的点积结果。（个人觉得这块也可以理解成给other添加了维度，然后再去掉此维度，只不过维度是(3, )而不是规则(3)中的( ,4)了，但是可能就是因为内部机制不同，所以官方说的是点积而不是维度的升高和下降）

```python
x = torch.rand(3) #1D
y = torch.rand(4,3) #2D
print(torch.matmul(y,x),'\n',torch.matmul(y,x).size()) #2D*1D
```

5. 如果一个维度至少是1D，另外一个大于2D，则返回的是一个批矩阵乘法（ a batched matrix multiply）。
  - 若input是1D，other是大于2D的，则类似于规则(3)

```python
x = torch.randn(2, 3, 4)
y = torch.randn(3)
print(torch.matmul(y, x),'\n',torch.matmul(y, x).size()) 
```

 - 若other是1D，input是大于2D的，则类似于规则(4)

```python
x = torch.randn(2, 3, 4)
y = torch.randn(4)
 
print(torch.matmul(x, y),'\n',torch.matmul(x, y).size())
```

```python
def use_svg_display():
    # Display in vector graphics
    display.set_matplotlib_formats('svg')

def set_figsize(figsize=(3.5, 2.5)):
    use_svg_display()
    # Set the size of the graph to be plotted
    plt.rcParams['figure.figsize'] = figsize

set_figsize()
plt.figure(figsize=(10, 6))
plt.scatter(features[:, 1].numpy(), labels.numpy(), 1)
```

```python
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    # The examples are read at random, in no particular order
    random.shuffle(indices)
    for i in range(0, num_examples, batch_size):
        j = torch.tensor(indices[i: min(i + batch_size, num_examples)])
        yield features[j], labels[j]
```

```python
batch_size = 10

for X, y in data_iter(batch_size, features, labels):
    print(X, y)
    break
```

### 初始化

我们将权重初始化成均值为0、标准差为0.01的正态随机数，偏差则初始化成0。

```python
w = torch.zeros(size=(num_inputs, 1)).normal_(std=0.01)
b = torch.zeros(size=(1,))
```

```python
w.requires_grad_(True)
b.requires_grad_(True)
```

requires_grad_(True)是自动求求梯度函数

### 线性模型定义

```python
def linreg(X, w, b):
    return torch.matmul(X, w) + b
```

### 损失函数定义

这里注意要使y的值维度转换为y^的维度

```python
def squared_loss(y_hat, y):
    return (y_hat - y.reshape(y_hat.shape)) ** 2 / 2
```

### 优化函数定义

在每个步骤中，使用从我们的数据集中随机抽取的一批数据，我们将根据我们的参数估计损失的梯度。然后，我们将在减少损失的方向上少量更新参数。假设已经计算了梯度，每个参数（param）都已经将其梯度存储在参数梯度. 下面的代码应用SGD更新，给定一组参数、学习率和批处理大小。更新步骤的大小由学习速率lr决定。因为我们的损失是作为一批示例的总和来计算的，所以我们用批大小（batch_size）来规范化步长，这样典型步长的大小不会严重依赖于我们对批大小的选择。

```python
def sgd(params, lr, batch_size):
    for param in params:
        param.data.sub_(lr*param.grad/batch_size)
        param.grad.data.zero_()
```

```python


lr = 0.03  # Learning rate
num_epochs = 3  # Number of iterations
net = linreg  # Our fancy linear model
loss = squared_loss  # 0.5 (y-y')^2

for epoch in range(num_epochs):
    # Assuming the number of examples can be divided by the batch size, all
    # the examples in the training data set are used once in one epoch
    # iteration. The features and tags of mini-batch examples are given by X
    # and y respectively
    for X, y in data_iter(batch_size, features, labels):
        l = loss(net(X, w, b), y)  # Minibatch loss in X and y
        l.mean().backward()  # Compute gradient on l with respect to [w,b]
        sgd([w, b], lr, batch_size)  # Update parameters using their gradient
    with torch.no_grad():
        train_l = loss(net(features, w, b), labels)
        print('epoch %d, loss %f' % (epoch + 1, train_l.mean().numpy()))

```

学习的结果

```python
print('Error in estimating w', true_w - w.reshape(true_w.shape))
print('Error in estimating b', true_b - b)
```
